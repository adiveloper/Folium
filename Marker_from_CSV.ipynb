```python
import folium
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Bryan1_waypoints.csv") 
df.head()
len(df)
```

```python
array = np.array(df)
lon = np.array(array[:,1])
lat = np.array(array[:,0])
print(lat[1:5],lon[1:5])
```

```python
# Make an empty map
m = folium.Map(location=[30.6719974283, -96.37366472], tiles="OpenStreetMap", zoom_start=40)
 
# I can add marker one by one on the map
for i in range(len(df)):
    if(i==1 and i!= len(df)):
        folium.Marker([lat[i],lon[i]],icon=folium.Icon(icon='cloud',color='red')).add_to(m)
    elif(i%8==0 and i!=808):
        folium.CircleMarker([lat[i],lon[i]],popup = None,).add_to(m)
m
```

```python
m.save("mymap.html")
```
